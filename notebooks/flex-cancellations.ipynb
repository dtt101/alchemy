```python
import pandas as pd

# load cancellations data and set the index to the date column
cancellations = pd.read_csv(
    '../data/cancellations.csv', 
    index_col='DATE', 
    parse_dates=True
)

# plot time series of COUNT_CANCELLATIONS with a separate line for each VARIATION_ID
cancel_pivot = cancellations.pivot(columns='VARIATION_ID', values='COUNT_CANCELLATIONS')
ax = cancel_pivot.plot()
```

```python
print(cancel_pivot.describe())
ax = cancel_pivot.boxplot()
```

```python
# compute correlation between '0' and '1' columns in cancel_pivot
print(cancel_pivot[[0, 1]].corr(method='spearman'))
```
